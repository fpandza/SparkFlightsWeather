```
%sql

drop table if exists weather_dates_icaos;

create table weather_dates_icaos as
select icao, date_id, min(temperature) as temperature, min(visibility) as visibility, min(wind_speed) as wind_speed, min(fog_flag) as fog_flag, min(rain_flag) as rain_flag, 
min(snow_flag) as snow_flag, min(hail_flag) as hail_flag, min(thunder_flag) as thunder_flag, min(tornado_flag) as tornado_flag
from (
  select distinct h.icao, 
  cast(trim(w.yearmoda) as int) as date_id, 
  cast(temp as double) as temperature, 
  cast(
  CASE WHEN trim(visib) = '999.9' THEN -1 ELSE visib END
  as double) as visibility, 
  cast(
  CASE WHEN trim(wdsp) = '999.9' THEN -1 ELSE wdsp END 
  as double) as wind_speed,
  cast(substr(frshtt, 2, 1) as integer) as fog_flag,
  cast(substr(frshtt, 3, 1) as integer) as rain_flag,
  cast(substr(frshtt, 4, 1) as integer) as snow_flag,
  cast(substr(frshtt, 5, 1) as integer) as hail_flag,
  cast(substr(frshtt, 6, 1) as integer) as thunder_flag,
  cast(substr(frshtt, 7, 1) as integer) as tornado_flag
  from weather w
  left outer join isd_history h
  on w.stn = h.usaf
  and w.wban = h.wban
  where h.icao is not null
) foo
group by icao, date_id;
```

```
%sql
drop table if exists flights_weather;

create table flights_weather as
select distinct f.*, temperature, visibility, wind_speed, fog_flag, rain_flag, snow_flag, hail_flag, thunder_flag, tornado_flag
from (
  select cast(concat(concat(replace(f.fl_date, '-', ''), f.OP_CARRIER_FL_NUM), f.CRS_DEP_TIME) as bigint) as trip_id,
  cast(replace(f.fl_date, '-', '') as int) as date_id, 
  --cast(concat(concat(concat(concat(concat(f.fl_date, ' '), substr(lpad(cast(f.dep_time as string), 6, '0'), 1, 2)), ':'), substr(lpad(cast(f.dep_time as string), 6, '0'), 3, 2)), ':00') as timestamp) as localdate, 
  cast(substr(lpad(cast(f.dep_time as string), 6, '0'), 1, 2) as int) as dep_hour,
  cast(substr(lpad(cast(f.arr_time as string), 6, '0'), 1, 2) as int) as arr_hour,
  cast(f.dep_delay as int), 
  cast(f.arr_delay as int), 
  cast(f.distance as int), 
  f.origin as src, 
  f.dest as dst, 
  o.city as city_src, 
  d.city as city_dst, 
  o.state as state_src, 
  d.state as state_dst
  from flights f 
  join airports o on o.iata = f.origin 
  join airports d on d.iata = f.dest
) f
join airports_dat ad on f.src = ad.iata 
join weather_dates_icaos wdi on ad.icao = wdi.icao and f.date_id = wdi.date_id
;
```

```
%sql
drop table if exists flights_weather_dist;

create table flights_weather_dist as

from (
  select cast(concat(concat(replace(f.fl_date, '-', ''), f.OP_CARRIER_FL_NUM), f.CRS_DEP_TIME) as bigint) as trip_id,
  cast(replace(f.fl_date, '-', '') as int) as date_id, 
  --cast(concat(concat(concat(concat(concat(f.fl_date, ' '), substr(lpad(cast(f.dep_time as string), 6, '0'), 1, 2)), ':'), substr(lpad(cast(f.dep_time as string), 6, '0'), 3, 2)), ':00') as timestamp) as localdate, 
  cast(substr(lpad(cast(f.dep_time as string), 6, '0'), 1, 2) as int) as dep_hour,
  cast(substr(lpad(cast(f.arr_time as string), 6, '0'), 1, 2) as int) as arr_hour,
  cast(f.dep_delay as int), 
  cast(f.arr_delay as int), 
  cast(f.distance as int), 
  f.origin as src, 
  f.dest as dst, 
  o.city as city_src, 
  d.city as city_dst, 
  o.state as state_src, 
  d.state as state_dst
  from flights f 
  join airports o on o.iata = f.origin 
  join airports d on d.iata = f.dest
) f
join airports_dat ad on f.src = ad.iata
distribute by f.date_id, ad.icao
--join weather_dates_icaos wdi on ad.icao = wdi.icao and f.date_id = wdi.date_id
;
```
